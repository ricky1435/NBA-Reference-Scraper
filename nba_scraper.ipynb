```python
# Library for opening url and creating
# requests
import urllib.request
    # pretty-print python data structures
from pprint import pprint
    # for parsing all the tables present
    # on the website
from html_table_parser.parser import HTMLTableParser
    # for converting the parsed data in a
    # pandas dataframe
import pandas as pd
import os
os.chdir('/Users/ruthvikpvs/Desktop/Projects/Nba and Aba Players')
def scraping_data(yes):
    import urllib.request
    # pretty-print python data structures
    from pprint import pprint
    # for parsing all the tables present
    # on the website
    from html_table_parser.parser import HTMLTableParser
    # for converting the parsed data in a
    # pandas dataframe
    import pandas as pd
    if yes =='TRUE':
        df = pd.DataFrame(columns = ['Player', 'From', 'To', 'Pos', 'Ht', 'Wt', 'Birth Date', 'Colleges']) # creates master dataframe
        letters ='abcdefghijklmnopqrstuvwxyz'
        for letter in letters:
            add_letter = str(letter) + '/'
            url = 'https://www.basketball-reference.com/players/' + add_letter
            req = urllib.request.Request(url=url)
            f = urllib.request.urlopen(req)
            f=f.read()
            xhtml =f.decode('utf-8')
            p = HTMLTableParser()

            # feeding the html contents in the
            # HTMLTableParser object
            p.feed(xhtml)
            d1=pd.DataFrame(p.tables[0])
            headers = d1.iloc[0]
            new_df  = pd.DataFrame(d1.values[1:], columns=headers)
            df=df.append(new_df)
        return df
```

## Division I,II and III data from NCAA College Men's Basketball website

```python
import urllib.request
    # pretty-print python data structures
from pprint import pprint
    # for parsing all the tables present
    # on the website
from html_table_parser.parser import HTMLTableParser
    # for converting the parsed data in a
    # pandas dataframe
import pandas as pd
```

```python
def college_players_div1():
    import urllib.request
    # pretty-print python data structures
    from pprint import pprint
    # for parsing all the tables present
    # on the website
    from html_table_parser.parser import HTMLTableParser
    # for converting the parsed data in a
    # pandas dataframe
    import pandas as pd
# DIVISION I PLAYERS
    list1 = ['p1','p2','p3','p4','p5','p6','p7','p8']

    data1 = pd.DataFrame(columns=['Rank', 'Name', 'Team', 'Cl', 'Height', 'Position', 'G', 'FGM', 'FT','PTS'])
    for i in list1:
        url = 'https://www.ncaa.com/stats/basketball-men/d1/current/individual/600' +'/'+ i
        req = urllib.request.Request(url=url)
        f = urllib.request.urlopen(req)
        f=f.read()
        xhtml =f.decode('utf-8')
        p = HTMLTableParser()

        # feeding the html contents in the
        # HTMLTableParser object
        p.feed(xhtml)
        d1=pd.DataFrame(p.tables[0])
        headers = d1.iloc[0]
        new_df1  = pd.DataFrame(d1.values[1:], columns=headers)
        data1=data1.append(new_df1)
    data1=data1.reset_index()
    data1.drop(columns=['index'],inplace=True,axis=1)
    for index,value in enumerate(data1.Rank):
        data1['Rank'][index]=index+1
    data1['division']='division1'
    return data1
```

```python
def college_players_div2():
    
    list2 = ['p1','p2','p3','p4','p5','p6']
    data1 = college_players_div1()
    data2 = pd.DataFrame(columns=['Rank', 'Name', 'Team', 'Cl', 'Height', 'Position', 'G', 'FGM', 'FT','PTS'])
    for i in list2:
        url = 'https://www.ncaa.com/stats/basketball-men/d2/current/individual/600' +'/'+ i
        req = urllib.request.Request(url=url)
        f = urllib.request.urlopen(req)
        f=f.read()
        xhtml =f.decode('utf-8')
        p = HTMLTableParser()

        # feeding the html contents in the
        # HTMLTableParser object
        p.feed(xhtml)
        d1=pd.DataFrame(p.tables[0])
        headers = d1.iloc[0]
        new_df1  = pd.DataFrame(d1.values[1:], columns=headers)
        data2=data2.append(new_df1)
    data2=data2.reset_index()
    data2['division']='division2'
    data2.drop(columns=['index'],inplace=True,axis=1)
    for index,value in enumerate(data2.Rank):
        data2['Rank'][index]=index+1
    data1=data1.append(data2)
    return data1
```
